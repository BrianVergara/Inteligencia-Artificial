```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generar datos de ejemplo
np.random.seed(0)
num_estudiantes = 200
horas_estudio = np.random.uniform(1, 10, num_estudiantes).reshape(-1, 1)  # Horas de estudio entre 1 y 10
```

```python
# Supongamos que queremos predecir las notas basadas en las horas de estudio
# Podemos generar las notas como una función lineal de las horas de estudio más un término de error aleatorio
coeficiente_verdadero = 0.5
intercepto_verdadero = 3.0
error = np.random.normal(0, 0.5, num_estudiantes)  # Término de error aleatorio
notas = coeficiente_verdadero * horas_estudio.squeeze() + intercepto_verdadero + error
# Limitar las notas al rango de 1 a 7
notas = np.clip(notas, 1, 7)
```

```python
# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(horas_estudio, notas)
```

```python
# Coeficientes del modelo
coeficiente = modelo.coef_
intercepto = modelo.intercept_
print("Coeficiente:", coeficiente)
print("Intercepto:", intercepto)
```

```python
# Predicción para todas las horas de estudio
horas_predichas = modelo.predict(horas_estudio)
```

```python
# Graficar los datos y la línea de regresión
plt.scatter(horas_estudio, notas, color='blue', label='Datos')
plt.plot(horas_estudio, horas_predichas, color='red', label='Regresión lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Notas')
plt.title('Regresión lineal: Horas de estudio vs Notas')
plt.legend()
plt.ylim(1, 7)  # Limitar el eje y al rango de notas de 1 a 7
plt.show()
```
